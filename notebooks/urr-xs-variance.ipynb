```julia
using DelimitedFiles
using Plots
using DataFrames
using DataFramesMeta
using LaTeXStrings
using Statistics
using Jammy
using Random,Distributions
```

```julia
Dav = 4
elam = [0.1:4.0:1.0e5;]
len = size(elam)[1]
d = Normal(0.05,0.02)
gg = rand(d,len)
d = Normal(0.004,0.002)
gn = rand(d,len)
jpi = ones(len)*4;

gg = broadcast(abs,gg)
gn = broadcast(abs,gn)
sqrtElam = broadcast(sqrt,elam)
gn = gn;
```

```julia
# read SAMMY calcs in
totlst = readdlm(joinpath(pwd(),"../data/ta181tot.lst"),Float64);
#### User input #############################
egrid::Vector{Float64} = 10 .^(range(-2,stop=6,length=5000))#totlst[:,1]#[0.01:0.0001:12.0;]
respars = DataFrame(
    Jpi = jpi,#[4.0,   3.0],
    Elam= elam,#[4.2801, 10.34],
    Gg  = gg,#[0.0530,0.0550],
    Gn  = gn,#[0.0039,0.00466],
)
# respars = DataFrame(
#     Jpi = [4.0],
#     Elam= [4.2801],
#     Gg  = [0.0530],
#     Gn  = [0.0039],
# )
targetspin = 3.5
targetmass = 180.94803
projmass = 1 # neutron 
projspin = 0.5
ac = 8.1271  # fermi = 1e-15 m
#############################################
start = time()
sigt,sign,sigx,sigc = Jammy.calcxs(egrid,respars,targetspin,targetmass,projspin,projmass,ac);
endt = time()
print(endt-start," seconds")
```

```julia
plotlyjs()
plot(egrid,sigt,
color="black",
linestyle=:solid,
label= L"\sigma_t(E)",
yaxis=:log,xaxis=:log,xlims=(1e0,1.e6),ylims=(1e0,1.0e5),legendfont=font(14),legend=:topleft,
xlabel="Energy [eV]",ylabel="Cross section [barns]")
```
